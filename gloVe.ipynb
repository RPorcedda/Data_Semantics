```python
from glove import Corpus, Glove
```

#GLOVE PAST 

```python
with open('/content/drive/MyDrive/DataSemantics/progetto/past_lem.txt', mode = 'r') as f:
  sentences_past_lem = [k.split() for k in f.readlines()]
```

**window**=5

```python
# creating a corpus object
corpus = Corpus() 
#training the corpus to generate the co occurence matrix which is used in GloVe
corpus.fit(sentences_past_lem, window=5)
#creating a Glove object which will use the matrix created in the above lines to create embeddings
#We can set the learning rate as it uses Gradient Descent and number of components
glove = Glove(no_components=100, max_loss=5)

glove.fit(corpus.matrix, epochs=5, no_threads=4, verbose=True)
glove.add_dictionary(corpus.dictionary)
glove.save('/content/drive/MyDrive/DataSemantics/progetto/models/GLOVE/glove_past_lem.model')
```

**window**=3

```python
# creating a corpus object
corpus = Corpus() 
#training the corpus to generate the co occurence matrix which is used in GloVe
corpus.fit(sentences_past_lem, window=3)
#creating a Glove object which will use the matrix created in the above lines to create embeddings
#We can set the learning rate as it uses Gradient Descent and number of components
glove = Glove(no_components=100, max_loss=5)
 
glove.fit(corpus.matrix, epochs=5, no_threads=4, verbose=True)
glove.add_dictionary(corpus.dictionary)
glove.save('/content/drive/MyDrive/DataSemantics/progetto/models/GLOVE/glove_past_lem_3w.model')
```

```python
past_lem = Glove.load('/content/drive/MyDrive/DataSemantics/progetto/models/GLOVE/glove_past_lem.model')
```

```python
past_lem.most_similar('greed', 15)
```

```python
past_lem_3 = Glove.load('/content/drive/MyDrive/DataSemantics/progetto/models/GLOVE/glove_past_lem_3w.model')
```

```python
past_lem_3.most_similar('pride', 15)
```
