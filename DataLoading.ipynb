Function that reads from the text file

```python
def read_text_file(file_path):
    with open(file_path, 'r') as f:
        body.append(f.readlines()[2:])
```

Lets iterate over all the desired file and read the body:

```python
import os
import glob
# Folder Path
path = PATH
  
# Change the directory
os.chdir(path)
  
# Read text File
body=[]
year=[]
category=[]
id=[]
for filename in glob.glob(os.path.join(PATH,'*','*.txt')):
    if filename.endswith(".txt"):
      id.append(filename.rpartition('_')[2].rpartition('.')[0])
      year.append(filename.rpartition('_')[0].rpartition('_')[2])
      category.append(filename.rpartition('_')[0].rpartition('_')[0][-2:])
      read_text_file(filename)
```

```python
import pandas as pd
df = pd.DataFrame(list(zip(id, category, year, body)),
               columns =['ID','Category', 'Year', 'Body'])
df.to_csv('coha.csv')
```
